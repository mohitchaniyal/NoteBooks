```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from collections import Counter
def counter(para):
    c=Counter()
    for i in para:
        c.update(i)
    return c
```

```python
df=pd.read_csv("../Data_sets/so-survey-2022/survey_results_public.csv") #https://insights.stackoverflow.com/survey
```

```python
lang_used=pd.DataFrame(counter(df.LanguageHaveWorkedWith.dropna().apply(lambda x:x.split(";"))).items(),columns=["language","user"]).sort_values("user")[-15:]
```

```python
plt.xkcd()
plt.barh(lang_used.language,lang_used.user,color="#e60049")
plt.xticks()
plt.title("Most Popular Programming Languages In 2022")
plt.xlabel("Popularity")
plt.tight_layout()
plt.savefig("MostPopularLang2022.png",bbox_inches='tight')
plt.show()
```
